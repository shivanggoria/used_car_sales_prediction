```python
from pyspark.ml.feature import StringIndexer, IndexToString, StandardScaler, Normalizer, MinMaxScaler, VectorIndexer, DenseVector, VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.regression import LinearRegression,GBTRegressor
```

## Imported important libraries for machine learning algoritmn and data profiling

```python
car_sales= ss.read.option('header','true').csv('newfile1.csv' , inferSchema=True)
```

## Imported csv file and converted it into pyspark dataframe

```python
car_sales.show(10)
```

```python
car_sales.printSchema()
```

```python
car_sales1=car_sales.drop("Car_name","Seats","State_Code","City_Code","Postal_Code","Mileage","_c0","Sales_ID")
```

## Removing unnecessary Columns

```python
car_sales=car_sales1 
car_sales.show() 
```

```python
car_sales.printSchema()
```

```python
strIndexer = StringIndexer(inputCols=["Brand_Name","Seller_Type","Fuel","Owner","City_Name","Transmission"], outputCols=["Brand_Name1","Seller_Type1","Fuel1","Owner1","City_Name1","Transmission1"])
```

```python
transformed_car_sales=strIndexer.fit(car_sales).transform(car_sales)
```

## Converting The columns with string values to integer so that we can apply ML model
## Transforming the converted data

```python
transformed_car_sales.printSchema()
type(transformed_car_sales)
```

```python
transformed_car_sales=transformed_car_sales.drop("Brand_Name","Seller_Type","Fuel","Owner","City_Name","transmission")
transformed_car_sales.show()
```

## Dropping duplicate columns with string values which were converted to integer type

```python
transformed_car_sales1 = transformed_car_sales.select(transformed_car_sales.columns[0],transformed_car_sales.columns[2], transformed_car_sales.columns[3],transformed_car_sales.columns[4],transformed_car_sales.columns[5],transformed_car_sales.columns[6],transformed_car_sales.columns[7],transformed_car_sales.columns[8],transformed_car_sales.columns[9],transformed_car_sales.columns[10],transformed_car_sales.columns[1])
transformed_car_sales1.show()
```

```python
transformed_car_sales1.printSchema()
```

```python
collectiveData = transformed_car_sales1.rdd.map(lambda x: (DenseVector(x[0:9]),x[10]))
```

```python
collectiveData.collect()
```

## Created dense vector in which first column has all features and second has ClassValue( Selling Price we want to predict)


```python
separatedData = ss.createDataFrame(collectiveData, ["feature","classLabels"])
```

```python
separatedData.show()
```

## Splitted the data into two columns named "features" and "classLabels"

```python
mxsc=MinMaxScaler(inputCol="feature",outputCol="sfeature")
```

```python
sp=mxsc.fit(separatedData).transform(separatedData)
```

```python
sp.show(5,truncate=False)
```

```python
sp1=sp.drop("feature")
```

```python
sp1.show(5,truncate=False)
```

## Applied Normalization to make data of same scale
## 0 to 1 Normalization is applied

```python
separatedData=sp1
separatedData.show(5)
```

```python
separatedData.printSchema()
```

```python
(trainingData, testingData) = separatedData.randomSplit([0.9,0.1])
```

## Splitting the data into traing set and test set

```python
print("Train Data :")
trainingData.show(5)
print("Test Data :")
testingData.show(5)
```

## Checking Training Data and Testing Data

```python
gbt = GBTRegressor(featuresCol = 'sfeature', labelCol = 'classLabels', maxIter=10)
```

## Applying Gradient-boosted tree regression ML Algoritmn

```python
gbt_model = gbt.fit(trainingData)
```

## Passing Training Data to ML Model

```python
gbt_predictions = gbt_model.transform(testingData)
```

## Testing Model against Testing Data

```python
gbt_predictions.select("classLabels","prediction").sample(0.01).show()
```

## Printing some random data from prediction table

```python
gbt_evaluator = RegressionEvaluator(labelCol="classLabels", predictionCol="prediction", metricName="r2")
rsquare = gbt_evaluator.evaluate(gbt_predictions)
print("R square test result = ", rsquare)
```

## R square test gives .96

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))

x_ax = range(0, gbt_predictions.count())
y_pred=gbt_predictions.select("prediction").collect()
y_orig=gbt_predictions.select("classLabels").collect()
 
plt.plot(x_ax, y_orig, label="original")
plt.plot(x_ax, y_pred, label="predicted")
plt.title("test data and predicted data")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='best',fancybox=True, shadow=True)
plt.grid(True)
plt.show() 
```
